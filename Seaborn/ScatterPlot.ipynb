# ScatterPlot

- Swarm plot Vs Scatter plot along with Violin plot
- JointPlot with some more keywords
- FacetGrid (advance version of catplot)
- PairGrid (advance version of Pairplot) Vs Pair plot
- KdePlot 1D (univariate) Vs 2D (Bivariate)
- RugPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
df.head(5)
```

###### FacetGrid (advance version of catplot)

```python
fg = sns.FacetGrid(df,row="sex",col="day",hue="time")
fg.map(sns.scatterplot,"total_bill","tip")
```

```python
fg = sns.FacetGrid(df,row="sex",col="day",hue="time")
fg.map(sns.regplot,"total_bill","tip")
```

###### PairGrid (advance version of Pairplot) Vs Pair plot

```python
sns.pairplot(df)
```

```python
pg = sns.PairGrid(df,hue="sex")
pg.map_upper(sns.regplot)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.lineplot)
```

```python
pg = sns.PairGrid(df,hue="sex")
pg.map_upper(sns.regplot)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.kdeplot)
```

###### KdePlot 1D (univariate) Vs 2D (Bivariate)

```python
df.head(2)
```

###### Univariate

```python
sns.kdeplot(df.total_bill)
```

###### Bivariate

```python
sns.kdeplot(df.total_bill,df.tip)
```

```python
sns.kdeplot(df.total_bill,df.tip,shade=True)
```

###### RugPlot

```python
sns.distplot(df.total_bill)
```

```python
sns.distplot(df.total_bill,kde=False)
```

```python
sns.distplot(df.total_bill,rug=True)
```

```python
sns.rugplot(df.total_bill)
sns.distplot(df.total_bill)
```
