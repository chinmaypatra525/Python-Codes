###### Voilin plot with and without inner data points

```python
plt.figure(figsize=(16,4))
sns.violinplot(mpg.mpg)
plt.show()
```

###### Add data points in violin plot

```python
plt.figure(figsize=(16,4))
sns.violinplot(mpg.mpg,inner="quartile") #"box", "quartile", "point", "stick",
sns.swarmplot(mpg.mpg,color="w",alpha=.3)
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.violinplot(mpg.mpg,inner="box",color="g") #"box", "quartile", "point", "stick",
sns.swarmplot(mpg.mpg,color="w",alpha=.3)
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.violinplot(mpg.origin,mpg.mpg)
sns.swarmplot(mpg.origin,mpg.mpg,color="c")
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.violinplot(mpg.origin,mpg.mpg,hue=mpg.cylinders)
# sns.swarmplot(mpg.origin,mpg.mpg,hue=mpg.cylinders)
plt.show()
```

###### Jointplot with various kind of margin plot

```python
sns.jointplot(mpg.acceleration,mpg.displacement)
```

```python
sns.jointplot(mpg.mpg,mpg.displacement)
```

```python
sns.jointplot(mpg.mpg,mpg.displacement,kind="reg",marginal_kws={"color":"r","kde":True},joint_kws={"logx":True})
```

###### HeatmapPlot

```python
a = np.random.randint(1,30,36).reshape(6,6)
a
```

```python
sns.heatmap(a)
```

```python
sns.heatmap(a,annot=True)
```

```python
sns.heatmap(a,annot=True,vmin=20,vmax=30)
```

```python
sns.heatmap(a,annot=True,fmt="2d",cbar=False)
```

```python
a = np.random.randint(1,30,36).reshape(6,6)
df1 = pd.DataFrame(a,columns=list("ABCDEF"))
df1
```

```python
corr = df1.corr()
corr
```

```python
sns.heatmap(corr,annot=True,fmt=".2f")
```

```python
sns.heatmap(corr,annot=True,fmt=".2f",vmin=.7,vmax=1.0)
```
